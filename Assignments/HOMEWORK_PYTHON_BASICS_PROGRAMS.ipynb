## Program 12: Factorial of N

**Problem**: Read `N` and print `N!`.

- Print factorial

**Sample Input**:
```

```
**Sample Output**:
```

```


````python
number = int(input("enter the factorial: "))
factorial = 1

if number < 0:
    print("Factorial is not defined for negative numbers")
elif number == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, number + 1):
        factorial *= i 
    print(f"The factorial of {number} is {factorial}")
````

## Program 28: Second Largest (Distinct)

**Problem**: Read `N` and a list of integers. Print the second largest **distinct** number.

**If second largest does not exist**, print: `Not Possible`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Second largest distinct value or `Not Possible`

**Sample Input**:
```
5
5 1 5 3 2
```
**Sample Output**:
```
3
```


````python
nums = input("enter the list: ").split()
largest = nums[0]
second = None
for x in nums[1:]:
    if x > largest:
        second = largest
        largest = x
    elif second is None or (x > second and x != largest):
        second = x
print("nums:", nums)
print("second_largest:", second)
````
